#### Metropolitan Areas(counties)

```python
#creating a 'counties' dataframe of all the Metropolitan areas from the dataframe
counties=df[df['Area Type']=='Metropolitan Area']

#grouping the areas along with the sum of Total job postings of each area
counties_series=counties.groupby("Area Name (County Names)")['Total  Job Openings'].sum()

#converting the above obtained pandas.Series into a dictionary 
counties_dict=counties_series.to_dict()
```

#### For x axis, we are coding the 29 Metropolitan areas from  1 to 29 as there is no space to show the complete names of the areas in the horizontal bar plot. The reference of the coded areas can be see right below the horizontal bar plot visualization.

```python
# x and y axis of the bar chart
bar_x=[str(i) for i in list(range(1,30))]  
bar_y=[round(i/1000000,2) for i in list(counties_dict.values())] #dividing it by 1,000,000 as the values are in millions

# Creating a plt.figure object to set the height and width of the output figure.
f = plt.figure()
f.set_figheight(15)
f.set_figwidth(11)

# Dividing the output into two subplots
plt.subplot(2,1,1)
plt.barh(bar_x,bar_y,height=0.7)
for index, value in enumerate(bar_y):
    plt.text(value, index,str(value),fontsize='small')
plt.ylabel("Metropolitan Area Name")
plt.xlabel("Total Job Openings(millions)")
plt.title("Total job openings in Metropolitan areas of California ",)
plt.show()

# Reference table for the county names
plt.subplot(2,2,1)
table_x=list(range(1,30))
table_y=list(counties_dict.keys())
table_data=[[table_x[i],table_y[i]] for i in range(len(table_x))]
table=plt.table(cellText=table_data, colLabels=['Index','Metropolitan Area'],colLoc='center', loc='center')
plt.axis('off')
#methods to set fontsize of the data in the table and the column widths(auto)
table.auto_set_font_size(False)
table.set_fontsize(9)
table.auto_set_column_width(col=[0,1]) #[0,1] are the indexes of the columns
plt.show()
```

###  Pie Chart

Pie Chart for the count of SOC levels in the Data set.

```python
pie=df.groupby('SOC Level')["SOC Level"].count()
pie_dict=pie.to_dict()
pie_dict
```

```python
#labels of the SOC levels - 1,2,3,4
labels = list(pie_dict.keys())

#count of SOC Levels
sizes = list(pie_dict.values())

#Exploding the first slice as it is the least value and almost not visible
explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Hogs')

#Pie charrt visualization with the percentage of the values
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=False, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### Line Plot

```python
#Each SOC Level with the sum of salries of jobs which belong to them stored as a dictionary
line=df.groupby('SOC Level')['Median Annual Wage'].sum()
line_dict=line.to_dict()
line_dict
```

```python
line_x=[str(item) for item in list(line_dict.keys())]
line_y=[item/1000000 for item in list(line_dict.values())]
plt.plot(line_x,line_y)
plt.xlabel('SOC Level')
plt.ylabel('Sum of salaries(millions)')
plt.title('Comparison of SOC Levels with their sum of salaries of respective jobs in that SOC level ')
plt.show()
```

###  Scatter Plot

```python
#Entry level education vs sum of salaries for the respective entry level education.
scatter=df.groupby('Entry Level Education')['Median Annual Wage'].sum()
scatter_dict=scatter.to_dict()

#removing 'No information available' since it is not valuable in the analysis
del scatter_dict['No information available']
scatter_dict
```

```python
scatter_x=[item/1000000 for item in list(scatter_dict.values())]  #df['Median Hourly Wage']   
scatter_y=list(scatter_dict.keys())  #df['Percentage Change']  

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55,])
plt.scatter(scatter_x,scatter_y,c=colors,cmap='viridis')
plt.colorbar()
plt.ylabel('Education Level')
plt.xlabel('Sum of Annual Salary(millions) for that level')
plt.title('Scatter plot to find correlation between education level and salary')
plt.show()
```
